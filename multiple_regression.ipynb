```
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```
df = pd.read_csv("/content/drive/My Drive/Colab Notebooks/BostonHousing.csv")
print(df.describe())
```

```
# First 13 columns as inputs
X=df.iloc[:,:-1].values

# Last columns as Target, which is price
Y=df.iloc[:,13].values

# Splitting dataset into trainnig and testing data, folllow the order of variables
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=4 )

# Creating Model
LR= LinearRegression()

# Fitting Model on training inputs i.e X_Train and training targets i.e Y_train 
LR.fit(X_train,Y_train)

# predict the house prices.
predictedPrice = LR.predict(X_test)
print(f"Predicted Price : {predictedPrice}")

# Mean square error 
print(f"Mean Square Error: {np.mean((predictedPrice - Y_test)**2)}")

#Percentage Accuracy
print(f"Percentage Accuracy: {round(LR.score(X_test,Y_test)*100)}%")

#Print the coefficient values of the regression object
print(f"Coefficient Values of the Regression Object: {LR.coef_}")
```
